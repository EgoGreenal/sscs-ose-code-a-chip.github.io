<h2>Introduction</h2>

<p>
Dynamic amplifiers such as <b>Ring Amplifiers</b>, <b>Inverter-Based Amplifiers</b>, and <b>Zero-Crossing Detector Amplifiers</b> have emerged as popular alternatives to small-signal settling-based amplifiers.
</p>


<p>
Unlike conventional OTAs (Operational Transconductance Amplifiers), where settling occurs in the small-signal and Islew domain<b></b>, dynamic amplifiers demonstrate <b>multiple stages of settling</b>, including:
</p>

<ul>
  <li><b>RC Settling</b></li>
  <li><b>Large Signal Settling</b></li>
  <li><b>Small Signal Settling</b></li>
</ul>

<p>
These mechanisms make dynamic amplifiers well-suited for high-performance, fast-settling analog circuits.
</p>

## Design of an Amplifier using Traditional $g_m/I_D$ Method

Amplifier design is typically done using the ($g_m/I_D$) based methodology. This approach provides an intuitive framework:

- In **strong inversion**, for a given overdrive voltage ($V_{ov}$), Designer can determine the transconductance ($g_m$).
- In **weak inversion**, Designer obtain a specific $g_m$ for a given drain current ($I_D$).

This methodology is especially useful since the design is based on **look-up tables**, making the process more streamlined and systematic.

---

### Application to Traditional Amplifiers

This design technique is well-suited for traditional **transconductance amplifier** design, where the **settling behavior is analyzed in the small-signal domain**.

---

### Application to Dynamic Amplifiers

So far, even **dynamic amplifiers** have been designed using the same ($g_m/I_D$) methodology to achieve a target **Bandwidth (BW)** which is, again, **small-signal in nature**.

However, there's a significant limitation:

- The **settling behavior** of dynamic amplifiers is **not strictly small signal**.
- We merely *hope* to get the best **$R_{on}$** for the **$g_m/I_D$** value chosen—originally selected for its small-signal bandwidth.
- There’s a level of uncertainty: we **hope** that the **non-linear settling behavior** contributes positively.
- But in reality, we **don't know** how much of the settling is being handled by the non-linear phase or **how effective** it is. This requires transient simulation iterations to fine tune the width and $R_{on}$

In short, we are **just hoping for the best**, without a clear understanding or control over the non-linear settling dynamics.


## $R_{on}/g_m$ - Based Design Methodology

In dynamic amplifiers, the settling behavior is influenced by both **non-linear** and **linear** settling phases. Each of these contributes significantly to the overall settling performance and must be accounted for during the design process.

---

### Non-Linear Settling Phase

- This is typically modeled as an **RC-type settling**, where an **impulse current** charges the load capacitor.
- The current during this phase is dictated by the **on-resistance** ($R_{on}$) of the transistor.
- This phase is critical for fast initial movement toward the final output voltage.

### Linear (Small-Signal) Settling Phase

- Once the output is closer to its final value, **small-signal settling** takes over.
- The behavior in this phase is governed by the **transconductance** ($g_m$).
- It determines the final fine settling and bandwidth of the amplifier.

---

### Design Motivation

The classic **$g_m/I_D$-based methodology** is powerful, but it comes with limitations:
- It does **not account for the signal swing requirements**.
- It does **not provide direct information about DC bias ranges**.
- It’s primarily tuned for small-signal operation and overlooks large-signal contributions.
- **$R_{on}/g_m$** based methodology should provide information regarding the voltage step that needs to be provided.
- **$R_{on}/g_m$** should provide additional information about Non-linear settling

To overcome this, we propose using the **$R_{on}/g_m$** ratio as a design-driven metric.

---

### $R_{on}/g_m$-Based Design Flow

- For a given *$R_{on}/g_m$*, we can quantify both:
  - The *large-signal settling speed* (via $R_{on}$),
  - And the *small-signal bandwidth* (via $g_m$).
  - Make a *unit cell* independent of C<sub>load</sub> design, but based on *voltage swing requirements*.

- A unit cell with fixed $I_D/W$ (i.e., fixed $g_m/I_D$) defines a base performance point.

- The required *$R_{on}$ across corners* for a given C<sub>load</sub> defines a *scaling multiplier* (width M).

- Scaling the unit cell by M gives:
  - $R_{on} \rightarrow R_{unit}/M$
  - $g_m \rightarrow g_{m,unit} \times M$
  - $I_D \rightarrow I_{D,unit} \times M$

- The *replica path* uses:
  - $R_{unit}$, $I_{unit}$, $W_{unit}$
  - Provides *replica tuning flexibility* through $I_{unit}$
  - *$g_m$ tuning* can be achieved by adjusting $I_{unit}$ and $W_{unit}$ across corners, without changing the main path width $W \times M$

---

### Key Insights

- Designer introduce *$R_{on}$ as a key design knob*, rather than focusing only on $g_m/I_D$.

- Enables *gm and voltage step tuning across corners* even after $R_{on}$ is fixed.

- Traditional *$g_m/I_D$-only design* ignores large-signal *$R_{on}$ based settling*.

- Poor $R_{on}$ choice can lead to *slow large-signal settling*, even with good $g_m$.

- Designer can have the same $g_m$ in three corners, even in *subthreshold* with fixed $I_D$,  
  ▸ but *$R_{on}$ may be much worse* in the slow corner — affecting settling.

- In this flow, Designer should *pick $R_{on}$ first*.

- Designer can *adjust $g_m$ later* using *replica bias* for the same $R_{on}$ and width.

- With *$R_{on}/g_m$ flow*, Designer:
  - Pick $R_{on}$ to ensure fast large-signal settling.
  - Then estimate how much settling time remains for small-signal (gm-based) behavior.

### Specifications / Design Goals

- $T_{settle} \approx 250$ ns (Approx. 20 MHz bandwidth for small-signal settling)
- $C_{total} \approx 1$ pF
- Determine ranges of $V_{GP}$, $V_{GN}$ for which the structure enters the small-signal settling domain

---

### Design of IBA using Traditional ($g_m/I_D$) Methodology

- For a 20 MHz bandwidth, the structure should provide:  
  $$
  G_m = 2\pi \cdot 20\,\text{MHz} \cdot C = 2\pi \cdot 20 \times 10^6 \cdot 1 \times 10^{-12} \approx 126\,\mu S
  $$

- Assume NMOS and PMOS contribute equally:  
  $$
  g_{mn} = g_{mp} = \frac{G_m}{2} \approx 64\,\mu S
  $$

- Let quiescent current $I_q = 1\,\mu A$ (assumed), used to generate $V_{DZN}$ and $V_{DZP}$

  Then:  
  $$
  g_{mn,dz} = g_{mp,dz} = 16\,\mu S
  $$  
  $$
  \frac{g_{mn,dz}}{I_q} = \frac{g_{mp,dz}}{I_q} = 16\,\frac{S}{A}
  $$

- This "deadzone" $G_m$ is then multiplied (using transistor multiplier circuits) to achieve the required $G_m$

> With the use of multipliers, we aim for an improved $R_{on}$, since $R_{on}$ reduces as the multiplication factor increases.

---

  - Assume $g_m/g_{ds} = 100$ (another assumption)
  - Use the model to estimate appropriate $(W/L)$ ratios

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import os

gmid_data = "./gmid_Data"

def get_data(data: list, key: str) -> pd.DataFrame:
    columns = ['Device', 'Length', 'Vgs', 'gm', 'id_val', 'Vth', 'gds', 'Cgg', 'Cgs', 'Cgd']
    labels = ['0.13u', '0.25u', '0.5u', '0.75u', '1u', '1.25u', '1.5u', '1.75u', '2u', '2.25u', '2.5u', '2.75u', '3u']
    new_data = []

    for i, file_path in enumerate(data):
        with open(f'{gmid_data}/{key}/{file_path}', 'r') as file:
            for line in file:
                entry = list(map(float, line.strip().split()))
                entry.insert(0, key)
                filtered_entry = [val for i, val in enumerate(entry) if i < 3 or i % 2 == 0]
                filtered_entry.insert(1, labels[i])
                new_data.append(filtered_entry)

    return pd.DataFrame(new_data, columns=columns)


files = {
    directory: sorted(os.listdir(os.path.join(gmid_data, directory)))
    for directory in sorted(os.listdir(gmid_data))
}


devices = ['HV_NMOS', 'HV_PMOS', 'LV_NMOS', 'LV_PMOS']
sel = 0 # select device here 0, 1, 2, 3

df = get_data(files[devices[sel]], devices[sel])

df['Vov'] = df['Vgs'] - df['Vth']
df['gm/id'] = df['gm'] / df['id_val']
df['gm/gds'] = df['gm'] / df['gds']
df['W'] = 2e-6
df['id/W'] = df['id_val'] / df['W']
df['ft'] = df['gm'] / (2 * np.pi * df['Cgg'])
df['Cgd/Cgg'] = df['Cgd'] / df['Cgg']
df['Cgs/Cgg'] = df['Cgs'] / df['Cgg']

# Convert 'Length' column to string for filtering consistency
df['Length'] = df['Length'].astype(str)

# Extract unique transistor lengths
length_labels = df['Length'].unique()
if sel < 2:
    length_labels = np.setdiff1d(df['Length'].unique(), ['0.13u', '0.25u'])

# Function to create an interactive Plotly graph with better scaling
def plot_interactive(df, x_col, y_col, x_label, y_label, title, log_y=False):
    fig = go.Figure()

    # Use a distinct color for each length
    colors = ['blue', 'red', 'green', 'purple', 'orange', 'brown', 'pink',
              'gray', 'olive', 'cyan', 'magenta', 'gold', 'lime']

    for i, length in enumerate(length_labels):
        length_df = df[df['Length'] == length]

        if length_df.empty:
            print(f"Warning: No data found for Length = {length}")
            continue

        fig.add_trace(go.Scatter(
            x=length_df[x_col],
            y=length_df[y_col],
            mode='lines',
            name=f"L={length}",
            line=dict(color=colors[i % len(colors)], width=2)
        ))

    # Auto-scale with scientific formatting
    fig.update_layout(
        title=title,
        xaxis_title=x_label,
        yaxis_title=y_label,
        legend_title="Transistor Length",
        hovermode="x unified",
        margin=dict(l=50, r=50, t=50, b=50),
        width=950,
        height=650
    )

    # Log scale for large values
    if log_y:
        fig.update_yaxes(type="log")

    # Format Y-axis to use SI notation (K, M, G, etc.)
    fig.update_yaxes(tickformat=".2s")  # ".2s" formats values like 1.2M, 3.4G

    # Auto-range for better scaling
    fig.update_xaxes(autorange=True, showgrid=True, zeroline=False)
    fig.update_yaxes(autorange=True, showgrid=True, zeroline=False)

    fig.show()


# Plot graphs with scaling improvements
plot_interactive(df, 'Vgs', 'gm/id', 'Vgs', 'gm/id', f'{df["Device"].unique()[0]} gm/id versus Vgs')
plot_interactive(df, 'Vov', 'gm/id', 'Vov', 'gm/id', f'{df["Device"].unique()[0]} gm/id versus Vov')
plot_interactive(df, 'gm/id', 'gm/gds', 'gm/id', 'gm/gds', f'{df["Device"].unique()[0]} gm/gds versus gm/id')
plot_interactive(df, 'gm/id', 'id/W', 'gm/id', 'id/W', f'{df["Device"].unique()[0]} id/W versus gm/id')  # Log scale for better visibility
plot_interactive(df, 'gm/id', 'ft', 'gm/id', 'ft (Hz)', f'{df["Device"].unique()[0]} ft versus gm/id')  # Log scale for frequency
plot_interactive(df, 'gm/id', 'Cgd/Cgg', 'gm/id', 'Cgd / Cgg', f'{df["Device"].unique()[0]} Cgd / Cgg versus gm/id')
plot_interactive(df, 'gm/id', 'Cgs/Cgg', 'gm/id', 'Cgs / Cgg', f'{df["Device"].unique()[0]} Cgs / Cgg versus gm/id')
```

```python
from scipy.optimize import minimize

# Target values
gm_id_target = 15
gm_gds_target = 120  # Must not be less than 100

# Weights (to control importance)
w1 = 1e7  # Weight for gm/id
w2 = 20  # Weight for gm/gds (more important)

# Objective function
def objective_function(x):
    adjusted_A, adjusted_B = x  # Values being optimized
    error_A = w1 * (adjusted_A - gm_id_target) ** 2
    error_B = w2 * (adjusted_B - gm_gds_target) ** 2
    return error_A + error_B  # Weighted total error

# Initial guess (mean values)
x0 = np.array([df['gm/id'].mean(), df['gm/gds'].mean()])

# Define bounds: (gm/id has no bounds, gm/gds must be ≥ 100)
bounds = [(None, None), (120, None)]  # No lower limit for gm/id, gm/gds >= 100

# Minimize using L-BFGS-B (supports bounds)
result = minimize(objective_function, x0, method='L-BFGS-B', bounds=bounds)

# Optimized values
optimized_gm_id, optimized_gm_gds = result.x

# Find the closest row in DataFrame
df['distance'] = np.sqrt(
    w1 * (df['gm/id'] - optimized_gm_id) ** 2 +
    w2 * (df['gm/gds'] - optimized_gm_gds) ** 2
)
closest_row = df.loc[df['distance'].idxmin()]  # Get the row with the smallest weighted distance

# Drop distance column after finding the closest row
df.drop(columns=['distance'], inplace=True)

# Print results
print("\nClosest Matching Row:\n", closest_row[['Device', 'Length', 'gm/id', 'gm/gds', 'id/W', 'ft']])
```

### Design of Inverter-Based Amplifier using the $\frac{R_{\text{on}}}{g_m}$ Methodology

#### Objective
To design an inverter-based amplifier targeting a **20 MHz bandwidth**, while ensuring effective settling behavior and low power consumption through the $\frac{R_{\text{on}}}{g_m}$ methodology.

---

#### 1. Transconductance Requirement
To achieve a bandwidth of $20\,\text{MHz}$, the required total transconductance is:
$$
G_m = 2\pi f C = 2\pi \cdot 20 \times 10^6 \cdot 1 \times 10^{-12} \approx 126\,\mu \text{S}
$$
Assuming equal contribution from NMOS and PMOS devices:
$$
g_{mn} = g_{mp} = \frac{G_m}{2} \approx 63\,\mu \text{S}
$$

---
#### 2. Target On-Resistance 
- To optimize **settling speed and power**, we allocate **50% of the settling** to the RC phase.
- This results in a target $R_{\text{on}} = 8\,k\Omega$, which must be the **maximum** across process corners to ensure robust **non-linear settling** behavior.
- We check $R_{\text{on}}$ vs Width to get the worst-case. The worst-case $R_{\text{on}}$ occurs at the **SS corner**, guiding our sizing choices.
From simulations:
- We find:
  $$
  \frac{R_{\text{on}}}{g_m} = 125 \times 10^6
  $$
  This implies:
  $$
  \frac{g_m}{I_D} \approx 17\,\text{V}^{-1} \quad 
  $$
  - TT: 0.658 V  
  - SS: 0.700 V
  
---
#### 3. Design Units and Scaling
- **Base Units Defined:**
  - $R_{\text{unit}} = 8\,k\Omega$
  - $I_{\text{unit}} = 1\,\mu A$
  - $g_{m,\text{unit}} = 17\,\mu \text{S}$
- The **main path** structure can be scaled using **multipliers** depending on application requirements (e.g., settling time, drive strength).
- The **replica path** can tune $g_m$ independently by adjusting only $I_{\text{unit}}$, leaving the main structure unchanged.

---
#### 4. Insights and Advantages
- This methodology offers deeper understanding of **non-linear settling** by explicitly incorporating $R_{\text{on}}$.
- Provides **predictable Deadzone regions**, eliminating the need for iterative simulations.
- Enables systematic tuning via transistor multipliers:
  - As the **multiplication factor increases**, $R_{\text{on}}$ decreases, further improving performance.

## Design of 5T-OTA Based on Swing Performance

In this section, we analyze the swing performance of a 5-transistor Operational Transconductance Amplifier (5T-OTA) and its application in setting deadzone voltages for Inverter based Amplifiers (IBAs).

The deadzone in the IBA, initially modeled using a Voltage-Controlled Voltage Source (VCVS), needs to be implemented using an OPAMP in negative feedback. These opamps should be capable of providing the voltage ranges necessary to place the IBA into its small-signal settling mode. 

The smaller the required deadzone range, the less complex the opamp needs to be. This enables simpler and more efficient designs.

We also define the range of input common-mode voltages and determine the **gm/ID** for each transistor that supports **maximum output common-mode swing**.

```python
def format_si(value, unit=""):
    """
    Convert a numerical value into a human-readable SI format.

    Args:
        value (float): The numerical value to convert.
        unit (str): The base unit (e.g., "s" for seconds, "W" for power).

    Returns:
        str: Formatted string with the appropriate SI prefix.
    """
    si_prefixes = {
        -12: "p",  # pico (p)
        -9: "n",   # nano (n)
        -6: "µ",   # micro (µ)
        -3: "m",   # milli (m)
        0: "",     # no prefix
        3: "k",    # kilo (k)
        6: "M",    # Mega (M)
        9: "G",    # Giga (G)
    }

    if value == 0:
        return f"0 {unit}"

    exponent = int(np.floor(np.log10(abs(value)) // 3) * 3)  # Get nearest multiple of 3
    exponent = max(min(exponent, 9), -12)  # Limit range between p (pico) and G (Giga)

    scaled_value = value / (10**exponent)
    si_prefix = si_prefixes.get(exponent, "")

    return f"{scaled_value:.3f} {si_prefix}{unit}"
```

```python
def get_data(file_path: str = 'Diff-amp_test_L0_13u.txt') -> pd.DataFrame:
  if use_with_cmfb:
    columns = ['Width', 'V(V1)', 'V(D1)', 'V(D2)', 'V(S)', 'gm', 'gds', 'VGS',
               'VTH', 'VDSAT', 'VDS', 'IDS', 'cgg']
    new_data = []
    current_width = 0  # Track the current width

    with open(file_path, 'r') as file:
        for line in file:
            entry = line.strip().split()

            if len(entry) == 24:  # If new width data is present
                current_width = float(entry[1])  # Second element is width
                entry = entry[2:]  # Remove width identifier and width itself
            entry.insert(0, current_width)  # Prepend width to the list

            # Keep only relevant columns (remove every alternate starting from index 3)
            filtered_entry = [val for i, val in enumerate(entry) if i < 3 or i % 2 == 0]

            new_data.append(filtered_entry)

    return pd.DataFrame(new_data, columns=columns).astype(float)
  else:
    columns = ['V(V1)', 'Width', 'V(D1)', 'V(D2)', 'V(S)', 'gm', 'gds', 'VGS',
               'VTH', 'VDSAT', 'VDS', 'IDS', 'cgg', 'gx']
    new_data = []

    with open(file_path, 'r') as file:
        for line in file:
            entry = line.strip().split()

            # Keep only relevant columns (remove every alternate starting from index 2)
            filtered_entry = [val for i, val in enumerate(entry) if i < 2 or i % 2 != 0]

            new_data.append(filtered_entry)

    return pd.DataFrame(new_data, columns=columns).astype(float)


lengths = [0.13e-6, 0.25e-6, 0.5e-6, 1.0e-6, 1.25e-6, 1.5e-6, 1.75e-6, 2.0e-6, 2.25e-6, 2.5e-6, 2.75e-6, 3.0e-6]

def format_length(L_m):
    L_um = L_m * 1e6
    if L_um.is_integer():
        return f"{int(L_um)}_0"
    elif (L_um * 10).is_integer():
        return f"{L_um:.1f}".replace('.', '_')  # e.g. 0.5 → 0_5
    else:
        return f"{L_um}".replace('.', '_')  # e.g. 1.25 → 1_25

data_files = {
    L: os.path.join(extract_path, f"Diff-amp_test_L{format_length(L)}u.txt")
    for L in lengths
} if use_with_cmfb else {
    L: os.path.join(extract_path, f"Diff_amp_wcmfb_L{format_length(L)}.txt")
    for L in lengths
}

cols = ['Length', 'Width', 'V(V1)', 'V(D1)', 'V(D2)', 'V(S)', 'gm', 'gds', 'VGS', 'VTH', 'VDSAT', 'VDS', 'IDS', 'cgg']

dfs = []
for length, file_path in data_files.items():
    df = get_data(file_path)
    df['Length'] = length
    dfs.append(df)

df = pd.concat(dfs, ignore_index=True)
df = df[cols]
```
